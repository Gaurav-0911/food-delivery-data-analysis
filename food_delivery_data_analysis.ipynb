```python
df["order_date"] = pd.to_datetime(df["order_date"])
```

```python
df["order_date"].head()
```

```python
df["month"] = df["order_date"].dt.month
```

```python
def get_quarter(month):
    if month in [1, 2, 3]:
        return "Q1 (JanMar)"
    elif month in [4, 5, 6]:
        return "Q2 (AprJun)"
    elif month in [7, 8, 9]:
        return "Q3 (JulSep)"
    else:
        return "Q4 (OctDec)"

df["quarter"] = df["month"].apply(get_quarter)
```

```python
df[["order_date", "month", "quarter"]].head()
```

```python
quarter_revenue = df.groupby("quarter")["total_amount"].sum()
quarter_revenue
```

```python
top_quarter = quarter_revenue.idxmax()
top_revenue = quarter_revenue.max()

top_quarter, top_revenue
```
